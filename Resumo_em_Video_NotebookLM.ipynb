# 🎬 Nova Funcionalidade: Resumos em Vídeo no NotebookLM

## Transforme qualquer tópico jurídico em vídeos educativos personalizados!

A mais nova adição ao NotebookLM permite criar vídeos educativos usando IA avançada, baseados nas 5.422+ questões reais do Firebase.

## ✨ **Principais Características**

### 🎯 **Formatos Disponíveis:**
- **Explicativo Didático**: Explicação detalhada e educativa
- **Resumo Rápido**: Conteúdo direto e conciso (2-5 min)
- **Aula Completa**: Apresentação abrangente (15-20 min)
- **Revisão para Prova**: Foco nos pontos mais cobrados

### ⏱️ **Durações Personalizáveis:**
- 2-5 minutos
- 5-10 minutos  
- 10-15 minutos
- 15-20 minutos

### 🎥 **Qualidades Disponíveis:**
- 720p HD
- 1080p Full HD  
- 4K Ultra HD

## 🚀 **Como Usar a Nova Funcionalidade**

### **Passo 1: Acesse a Nova Aba**
1. Abra o NotebookLM: `http://localhost:3003/notebook-ia`
2. Clique na nova aba **"Resumo em Vídeo"** (ícone de vídeo vermelho)

### **Passo 2: Configure seu Vídeo**
1. **Tópico Principal**: Digite o tema que deseja estudar
   - Exemplo: "Direito Constitucional - Princípios Fundamentais"
2. **Detalhes Específicos**: Adicione pontos que devem ser abordados
3. **Duração**: Escolha entre 2-20 minutos
4. **Formato**: Selecione o estilo do vídeo
5. **Qualidade**: Defina a resolução desejada

### **Passo 3: Gere seu Vídeo**
1. Clique em **"Gerar Resumo em Vídeo"**
2. Acompanhe o progresso em tempo real
3. Aguarde a conclusão do processamento

## 🧠 **Tecnologia Por Trás da Funcionalidade**

### **Integração com Firebase:**
- ✅ Extração automática de palavras-chave do tópico
- ✅ Busca inteligente nas 5.422+ questões jurídicas reais
- ✅ Identificação de matérias e conceitos relevantes
- ✅ Scoring de relevância baseado no conteúdo

### **Geração Inteligente de Conteúdo:**
```javascript
// Exemplo do processo de geração
1. Análise do tópico → Extração de palavras-chave
2. Consulta no Firebase → Busca questões relevantes  
3. Criação do roteiro → Script personalizado
4. Geração do vídeo → IA avançada
5. Processamento final → Vídeo pronto!
```

### **API Endpoint:**
- **URL**: `/api/notebook-ia/video-summary`
- **Método**: POST
- **Integração**: Firebase + IA de Vídeo
